```python
import glob
import cv2
import os
```

```python
def hour_video(filename):
    return int(filename.split('_')[-5][:2])
```

```python
folder = "./DATASET/results/10-10-2024/"
# folder = ".\\DATASET\\results\\10-10-2024\\"
folder = "./DATASET/lopes_videos_antigos/*/"
files = glob.glob(folder + "*cam_0.avi")
files = [f for f in files if hour_video(f) >5 and hour_video(f)<18]
files[::40].__len__()
```

```python
save_folder = "./DATASET/balls_segmentation/"
os.makedirs(save_folder,exist_ok=True)
hour_previous = 0
for f in files:
    if hour_video(f) == hour_previous:
        continue

    cap = cv2.VideoCapture(f)
    ret, frame = cap.read()

    if frame.shape[0] == 1080:
        filename = f.split('\\')[-1].split('.')[0]
        # print(filename)
        saved = cv2.imwrite(save_folder+filename+".png", frame)
        print(filename, saved)
        hour_previous = hour_video(f)
```
